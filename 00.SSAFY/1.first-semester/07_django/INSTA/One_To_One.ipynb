### OneToOne model

#### User

| id      | email     | password  |
| ------- | --------- | --------- |
| PK, INT | CharField | CharField |

#### Profile

| id      | user_id    | first_name | last_name |
| ------- | ------- | ---------- | --------- |
| PK, INT | FK, User, INT | CharField  | CharField |


```python
user = User.objects.create(email='sdnfosd@naver.com', password='123123')
```

```python
user
```

```python
user = User.objects.first()

profile = Profile()
profile.user = user
profile.first_name = 'TaeYoung'
profile.last_name = 'Yu'
profile.save()
```

```python
profile.user_id
```

```python
profile.user
# profile.user <==> User.objects.get(id=profile.user_id)
```

```python
u = user.profile.last_name
u
```

```python
Profile.objects.create(user=user, first_name='virus', last_name='Bad')
```

```python
Profile.objects.create(user_id=user.id, first_name='virus', last_name='Bad')
```

```python
User.objects.create(email='kim@ssafy.com', password='123123')
```

```python
new_user = User.objects.get(id=2)
Profile.objects.create(user=new_user, user_id=new_user.id, first_name='kim', last_name='ssafy')
```

```python
new_profile = Profile.objects.get(id=2)
new_profile.user_id
```
